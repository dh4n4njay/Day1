```python
from rdkit import Chem
from rdkit.Chem import Draw
mol = Chem.MolFromSmiles('CCCC')
# Display the molecule
img = Draw.MolToImage(mol)
display(img)
```

```python
mol2 = Chem.MolFromSmiles('CC=CC')
img2 = Draw.MolToImage(mol2)
display(img)
```

```python
mol3=Chem.MolFromSmiles('CCC#N')
img3= Draw.MolToImage(mol3)
display(img3)
```

```python
mol4 = Chem.MolFromSmiles('C[Br]')
img4 = Draw.MolToImage(mol4)
display(img4)
```

```python
mol_branch1=Chem.MolFromSmiles('CC(C)CC(=O)CC(#N)')
img_branch1=Draw.MolToImage(mol_branch1)
display(img_branch1)
```

```python
mol_n=Chem.MolFromSmiles('c1c(N(=O)=O)cccc1')
img_n=Draw.MolToImage(mol_n)
display(img_n)
```

## **Day2: Rings and Charged Molecules/Atoms**

```python
mol_ring=Chem.MolFromSmiles('c1ccccc1')
img_ring=Draw.MolToImage(mol_ring)
display(img_ring)
```

```python
mol_ring2=Chem.MolFromSmiles('c1cc2ccccc2cc1')
img_ring2=Draw.MolToImage(mol_ring2)
display(img_ring2)
```

```python
mol_ring3=Chem.MolFromSmiles('C1OC1CC')
img_ring3=Draw.MolToImage(mol_ring3)
display(img_ring3)
```

```python
mol_charge=Chem.MolFromSmiles("[n+]1ccccc1CC(=O)")
img_charge=Draw.MolToImage(mol_charge)
display(img_charge)
```

```python
mol_charge2=Chem.MolFromSmiles('CCC(=O)[O-]')
img_charge2=Draw.MolToImage(mol_charge2)
display(img_charge2)
```
