```python
from rdkit import Chem
from rdkit.Chem import Draw
mol = Chem.MolFromSmiles('CCCC')
# Display the molecule
img = Draw.MolToImage(mol)
display(img)
```

```python
mol2 = Chem.MolFromSmiles('CC=CC')
img2 = Draw.MolToImage(mol2)
display(img)
```

```python
mol3=Chem.MolFromSmiles('CCC#N')
img3= Draw.MolToImage(mol3)
display(img3)
```

```python
mol4 = Chem.MolFromSmiles('C[Br]')
img4 = Draw.MolToImage(mol4)
display(img4)
```

```python
mol_branch1=Chem.MolFromSmiles('CC(C)CC(=O)CC(#N)')
img_branch1=Draw.MolToImage(mol_branch1)
display(img_branch1)
```
